### **4. Cargar y Preparación de la Data**
En esta sección se definen los generadores de imágenes utilizando `ImageDataGenerator` de Keras, los cuales se encargarán de:

- Leer las imágenes desde las carpetas correspondientes
- Aplicar normalización de píxeles
- Realizar *data augmentation* (rotaciones, desplazamientos, zoom y flips) para mejorar la capacidad de generalización del modelo
- Dividir el conjunto en subconjuntos de entrenamiento (80%) y validación (20%)


```python
import os
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.metrics import classification_report, confusion_matrix
from statistics import mean, stdev
```

```python
# Configuración
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
INPUT_SHAPE = (128, 128, 3)
EPOCHS = 5
```

```python
# Ruta al dataset balanceado (preprocesado)
path_dataset = Path("dataset/resized")
```

#### **4.1 Generador de Datos con Keras**
En esta sección se define el generador de datos con Keras, el cual se encargará de:
- Leer las imágenes desde las carpetas.
- Aplicar normalización y data augmentation para mejorar la generalización del modelo.
- Dividir el conjunto en entrenamiento y validación de forma estratificada.

```python
# Generador de datos
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)
```

```python
train_generator = datagen.flow_from_directory(
    path_dataset,
    target_size=IMG_SIZE,          
    batch_size=BATCH_SIZE,
    class_mode='binary',            
    subset='training',
    shuffle=True,
    seed=42
)
```

```python
val_generator = datagen.flow_from_directory(
    path_dataset,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation',
    shuffle=False
)
```

### **5. Implementación del Modelo CNN**

A continuación se define la arquitectura de la red neuronal convolucional. Esta consiste en:

- Dos bloques de convolución + max-pooling
- Una capa `Flatten` para aplanar los mapas de características
- Un `Dropout` para reducir sobreajuste
- Dos capas densas, la última con activación sigmoide para clasificación binaria


```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=INPUT_SHAPE),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')  # esta es la salida binaria
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Entrenamiento
history = model.fit(train_generator, validation_data=val_generator, epochs=EPOCHS)
```

#### **5.1 Visualización del historial de entrenamiento**
A continuación se visualiza la evolución del *loss* y *accuracy* durante el entrenamiento y validación. Esto permite observar si el modelo ha convergido adecuadamente y si existe indicio de sobreajuste.


```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Pérdida durante el entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Precisión durante el entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

### **6. Medición de Tiempos e Inferencia**

Se realiza el proceso de inferencia sobre el conjunto de validación completo, repitiendo la medición cinco veces para calcular las siguientes métricas:

- Tiempo máximo
- Tiempo mínimo
- Tiempo promedio
- Desviación estándar

Además, se evalúa el rendimiento mediante la métrica de *accuracy* y el *reporte de clasificación*.


```python
# Tiempo de inferencia en 50% y 100% de los datos (validación)
y_true = val_generator.classes
y_pred_probs = []
tiempos = []

# Inferencia por lote
for _ in range(5):
    start = time.time()
    val_generator.reset()
    preds = model.predict(val_generator, verbose=0)
    tiempos.append(time.time() - start)
    y_pred_probs.append(preds)
```

```python
# Resultados estadísticos
tiempos_segundos = [round(t, 4) for t in tiempos]
print(f"\nTiempos de inferencia (5 repeticiones): {tiempos_segundos}")
print(f"Promedio: {mean(tiempos_segundos):.4f}s - Desv. Estándar: {stdev(tiempos_segundos):.4f}s")
```

```python
# Predicción final
preds_final = np.round(np.mean(y_pred_probs, axis=0)).flatten()
print("\nReporte de Clasificación:")
print(classification_report(y_true, preds_final))
```

### **7. Ejemplos de Resultados**
A continuación se visualizan 10 ejemplos de imágenes clasificadas, mostrando la clase real y la predicción del modelo. Esto permite verificar visualmente el desempeño y detectar posibles errores en las predicciones.


```python
val_generator.reset()
imgs, labels = next(val_generator)
preds_sample = model.predict(imgs)

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle("Ejemplos de predicciones")
for i, ax in enumerate(axes.flat):
    ax.imshow(imgs[i])
    pred = "Healthy" if preds_sample[i] < 0.5 else "Early_Blight"
    true = "Healthy" if labels[i] == 0 else "Early_Blight"
    ax.set_title(f"Pred: {pred}\nTrue: {true}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```
